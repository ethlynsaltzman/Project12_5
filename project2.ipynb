```python
#In this project I chose to compare the lilklihood of readmission to the hospital by race (specifically I was intrested to compare the readmission of African Americans with people of other races). Comparison was done seprately for men and women
# I also decided to look ta if there was a relationship between time in hospital
import pandas as pd
dfH = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
dfH.head(10)
```

```python
# some general descriptives of the diabetes data
dfH.describe()
```

```python
dfH.shape
```

```python
# filtered the data to the variables that intrested me and looke at the type of data types
dfH = dfH[['encounter_id', 'race','readmitted','gender','age','time_in_hospital']]
dfH.dtypes
```

```python
# changed time in hospital to float so could do computation with it
dfH['time_in_hospital'].apply(float)
```

```python
# I  wanted to check if there were any values that needed to be cleaned from this column
dfH['time_in_hospital'].unique()
```

```python
# I  wanted to check if there were any values that needed to be cleaned from this column
dfH['race'].unique()
```

```python
# Removed rows not identified with race as this was this group comparison I wanted to make
dfH = dfH.loc[dfH["race"] !=  "?" ]
```

```python
# I  wanted to check if there were any values that needed to be cleaned from this column
dfH['gender'].unique()
```

```python
# removed patients wirthout identified gender.
dfH = dfH.loc[dfH["gender"] !=  "Unknown/Invalid" ]
```

```python
# checking values for this column
dfH['readmitted'].unique()
```

```python
# decided to collapse into two categories (readmitted within 30 days and not readmitted within 30) I changed  0, 1  so I could easily get descriptives graphs that included the proption redmitted within 30 days
dfP = dfH.replace('NO', '0')
dfM = dfP.replace('>30','0')
dfN = dfM.replace('<30','1')
dfN.head()
```

```python
# changed the data type to numerical for readmisison
dfN['readmitted']= dfN['readmitted'].apply(float)

dfN = dfN[['encounter_id', 'race','readmitted','gender','age','time_in_hospital']]
```

```python
#created bar polt to compare chance of readmission by race and gender
import seaborn as sns

ax = sns.barplot(x = 'race', y = 'readmitted',data = dfN, hue = 'gender')
ax.bar_label(ax.containers[0])
```

```python
# there didnt seem to be much difference by race so I deceided to also look if there was atrelationship between time in the hospital and rate of readmission,
#I had some issue plotting after using  the group by function
#it created a df without a column for time in the hospital so I had to set time in hospital as not being in the index
dfr = dfN.groupby(['time_in_hospital'], as_index = False).agg ({'readmitted': 'mean'})
print(dfr)
```

```python
#plotted this relationsip
sns.regplot(x="time_in_hospital", y="readmitted", data=dfr);
```

```python
#  wanted to see if the correlation was significant
from scipy.stats.stats import pearsonr
pearsonr(dfr['time_in_hospital'], dfr['readmitted'])
```

The conclusion from this analysis. There appeared to be no significant difference between people in different racial groups in the rate of readmission. There is significant relationship between time in the hospital and readmission. With nmore days in the hospital increasing the readmission rate
